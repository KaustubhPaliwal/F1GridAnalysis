```python
import pandas as pd

# Load the datasets

f1_circuits_df = pd.read_csv('F1Data/F1-circuits.csv', encoding='utf-8', sep=';')
circuits_df = pd.read_csv('F1Data/circuits.csv')
```

```python
# Display the first few rows of each dataset
f1_circuits_df
```

```python
circuits_df
```

```python
f1_circuits_df = f1_circuits_df[['Circuit']]
f1_circuits_df
```

```python
circuits_df = circuits_df[['name']]
circuits_df
```

```python
# Extract unique values from both DataFrames
f1_unique = f1_circuits_df['Circuit'].unique()
circuits_unique = circuits_df['name'].unique()

# Find matching values
matching = set(f1_unique) & set(circuits_unique)

# Find non-matching values in f1_circuits_df
non_matching_f1 = set(f1_unique) - set(circuits_unique)

# Find non-matching values in circuits_df
non_matching_circuits = set(circuits_unique) - set(f1_unique)

# Display the results
print(matching)
print(non_matching_f1)
print(non_matching_circuits)
```
